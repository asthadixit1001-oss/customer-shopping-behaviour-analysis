Importing and Cleaning Data

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\Astha\OneDrive\Desktop\customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(r'\s+', '_', regex=True)
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchased_amount'})
```

```python
df.columns
```

Creating a Column age_group

```python
labels = ['Young Adult', 'Adult', 'Middle-Age', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels= labels)
```

```python
df.head(10)
```

```python
df.tail(10)
```

Create column purchased_frequency_days

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchased_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
